```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

network_df = pd.read_csv("SENA_Dataset - MCUActorList.csv") 
```

```python
edges = [tuple(x) for x in network_df[['Name','Actor_1']].values.tolist()]
B = nx.Graph()
B.add_nodes_from(network_df['Name'].unique(), bipartite=0)
B.add_nodes_from(network_df['Actor_1'].unique(), bipartite=1)

```

```python
for row in edges:
    B.add_edge(row[0], row[1])

left_or_top = network_df['Name'].unique()
pos = nx.bipartite_layout(B, left_or_top)

# Draw the graph
nx.draw(B,pos,node_color='#A0CBE2',edge_color='#00bb5e',width=1,
     edge_cmap=plt.cm.Blues,with_labels=True)

# Get the edge labels for ratings
edge_labels = nx.get_edge_attributes(B, "connect")

# Draw the edge labels
nx.draw_networkx_edge_labels(B, pos, edge_labels=edge_labels)
```

```python
# First specify the nodes you want on left or top
left_or_top = network_df['Name'].unique()

# Then create a bipartite layout
pos = nx.bipartite_layout(B, left_or_top)

# Pass that layout to nx.draw
nx.draw(B,pos,node_color='#A0CBE2',edge_color='#00bb5e',width=1,
     edge_cmap=plt.cm.Blues,with_labels=True)
```
